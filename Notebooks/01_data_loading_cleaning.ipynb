## Fuel Price Trends & Cost-Saving Analysis - Western Australia

```python
import pandas as pd
```

Load data from https://www.fuelwatch.wa.gov.au/retail/historic

```python
data1 = pd.read_csv('../Data/Raw_Data/FuelWatchRetail-11-2025.csv')
data2 = pd.read_csv('../Data/Raw_Data/FuelWatchRetail-12-2025.csv')
data3 = pd.read_csv('../Data/Raw_Data/FuelWatchRetail-10-2025.csv')
```

```python
data = pd.concat([data1, data2, data3], ignore_index=True)
```

```python
data.to_csv('../Data/Raw_Data/fuel_data_oct_nov_dec_2025.csv', index=False)
```

```python
data
```

```python
data.info()
```

Rename Columns - This improves readability of the dataset

```python
data.columns = [
    "publish_date",
    "trading_name",
    "brand",
    "fuel_type",
    "price",
    "address",
    "suburb",
    "postcode",
    "area",
    "region"
]
```

```python
data.head()
```

Check data types and change them accordingly

```python
data['publish_date'] = pd.to_datetime(data['publish_date'], format='%d/%m/%Y')
```

```python
data.info()
```

```python
data.fuel_type.unique()
```

Select ONE fuel type

```python
data.fuel_type.value_counts()
```

ULP has the most available data. Therefore let's select ULP for our analysis.

```python
df = data[data["fuel_type"] == "ULP"]
df.head()
```

```python
df.price.describe()
```

Create Time Features

```python
df['day'] = df['publish_date'].dt.day
```

```python
df['month'] = df['publish_date'].dt.month_name()
```

```python
df['day_name'] = df['publish_date'].dt.day_name()
```

```python
df
```

```python
df['weekend'] = df['day_name'].isin(['Saturday', 'Sunday'])
```

```python
df
```

```python
df.to_csv('../Data/Processed_Data/fuel_data_oct_nov_dec_2025.csv', index=False)
```
